## 1. Create a list‐of‐lists representation of this dataset (ignore the column headings).

```python
# Define a "ShowData" function - note the default value for the (now) optional parameter.
#  dataset is a list of lists
def ShowData(dataset = [['']]):
    for r in dataset:
        print (r)
        
# Define a list for the data.  Will be a list of lists.
data = []
# open the file
fname = "../data/stocks.csv"
f = open(fname, "r")

#Read lines, note that range starts from 2 (skips column header)
for i in range(2):
    line = f.readline()
# loop until we run out of lines

while (line):
    # strip the newline and tokenize (split on commas, in this case)
    tokens = line.rstrip().split(',')
    # append this record to the dataset
    data.append(tokens)
    # read the next line
    line = f.readline()
# close the file
f.close()
# show the data
ShowData (data)
```

## Day 1a - with csv export

```python
#redefine varible with new csv file 'stocksA'
stocks = pd.read_csv(("../data/stocksA.csv"), nrows=1)

#new_closing variable with new csv file 'newdayA'
    #nrows remains at 10 so the 10th day closing price from the day prior is dropped
new_closing = pd.read_csv("../data/newdayA.csv", names=["DAL Close", "HON Close", "DKNG Close", "WFC Close"],)
#old_closing variable is defined by the old dataset, named 'closingprice.csv'
old_closing = pd.read_csv("../data/closingprice.csv", names=["DAL Close", "HON Close", "DKNG Close", "WFC Close"],)

#redefine dataframe 'closing' combining old and new closing prices from 1-10 days
closing=pd.concat([new_closing, old_closing], sort=False, axis=0, ignore_index=True).loc[0:9]

#export closing df to overwrite closingprice.csv with updated values
closing.to_csv (r'../data/closingprice.csv', index = False, header=False)
    
    #redefine prices by concating updated stocks with new closing price datafame
prices=pd.concat([stocks, closing], sort=False, axis=1)

prices
```
