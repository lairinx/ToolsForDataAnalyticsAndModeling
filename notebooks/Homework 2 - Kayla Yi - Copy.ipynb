# Question 3
#### Modified data.append to include only 4 selected tokens. Defined each token type.

```python
# Initial version - "standard programming"
#Define ShowData function
def ShowData(dataset = [["No dataset sent"]]):
    for r in dataset:
        # print elements in a tab-separated format
        print ("\t".join(r))
# Define a list for the data.  Will be a list of lists.
data = []
# open the file
fname = "14_Foreign_Exchange_Rates_PureNumeric.csv"
f = open(fname, "r")
# ignore the first 5 lines
for i in range(6):
    line = f.readline()
# loop until we run out of lines
while (line):
    # strip the newline and tokenize (split on commas, in this case)
    tokens = line.rstrip().split(',')
    # append this record to the dataset
    data.append([(tokens[0]), tokens[1], (tokens[2]), (tokens[14])])
    tokens[0]=int(tokens[0])
    tokens[1]=str(tokens[1])
    tokens[2]=float(tokens[2])
    tokens[14]=float(tokens[14])
    # read the next line
    line = f.readline()
# close the file
f.close()
# show the data
ShowData(data)

#confirmation of item type
type(tokens[0]), type(tokens[1]), type(tokens[2]), type(tokens[14])
```

# Question 4
#### Modifed data.append with str.format command

```python
# Initial version - "standard programming"
#Define ShowData function
def ShowData(dataset = [["No dataset sent"]]):
    for r in dataset:
        # print elements in a tab-separated format
        print ("\t".join(r))
# Define a list for the data.  Will be a list of lists.
data = []
# open the file
fname = "14_Foreign_Exchange_Rates_PureNumeric.csv"
f = open(fname, "r")
# ignore the first 5 lines
for i in range(6):
    line = f.readline()
# loop until we run out of lines
while (line):
    # strip the newline and tokenize (split on commas, in this case)
    tokens = line.rstrip().split(',')
    # append this record to the dataset, modified data.append str.format
    tokens[0]=int(tokens[0])
    tokens[1]=str(tokens[1])
    tokens[2]=float(tokens[2])
    tokens[14]=float(tokens[14])
    data.append(["{}, {}, {:,.2f}, {:,.2f}".format(int(tokens[0]), tokens[1],float(tokens[2]), float(tokens[14]))])
    # read the next line
    line = f.readline()
# close the file
f.close()
# show the data
ShowData(data)

#confirmation of type for each item
type(tokens[0]), type(tokens[1]), type(tokens[2]), type(tokens[14])
```

#### UK list

```python
# Initial version - "standard programming"
#
# Define a list for the data.  Will be a list of lists.
data = []
ds=[]
# open the file
fname = "14_Foreign_Exchange_Rates_PureNumeric.csv"
f = open(fname, "r")
# ignore the first 5 lines
for i in range(6):
    line = f.readline()
# loop until we run out of lines
while (line):
    # strip the newline and tokenize (split on commas, in this case)
    tokens = line.rstrip().split(',')
    # append this record to the dataset
    data.append(tokens)
    ds.append([float(tokens[5])])
    # read the next line
    line = f.readline()
# close the file
f.close()
# define list
l=ds

#show list
l
```

#### Mean of United Kingdom data

```python
sum = 0
# iterate from first row to last row
for i in data:
    # add United Kingdom data of every row to sum
    sum = sum + float(i[5])
#define mean
mean = sum/len(ds)
mean
```
